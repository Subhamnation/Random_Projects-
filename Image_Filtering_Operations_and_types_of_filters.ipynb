```python
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
from PIL import ImageFilter
from PIL import Image
```

```python
I = imread("image23.jpg")
figure(figsize =(12, 10))

subplot(3, 4, 1)
plt.imshow(I)
title("Original")

subplot(3, 4 ,2)
I_1 = Image.open("image23.jpg")
I_2 = I_1.filter(ImageFilter.CONTOUR)
plt.imshow(I_2)
title("Contour")

subplot(3, 4 ,3)
#I_1 = Image.open("nature1.jpg")
I_3 = I_1.filter(ImageFilter.DETAIL)
plt.imshow(I_3)
title("Detail")

subplot(3, 4 ,4)
#I_1 = Image.open("nature1.jpg")
I_4 = I_1.filter(ImageFilter.EDGE_ENHANCE)
plt.imshow(I_4)
title("Edge enhanced")   # Edge enhanced filter is a laplacian filter. Basically it is convolution with laplacian function

subplot(3, 4 ,5)
#I_1 = Image.open("nature1.jpg")
I_5 = I_1.filter(ImageFilter.EDGE_ENHANCE_MORE)
plt.imshow(I_5)
title("Edge enhanced more")

subplot(3, 4 ,6)
#I_1 = Image.open("nature1.jpg")
I_6 = I_1.filter(ImageFilter.EMBOSS)
plt.imshow(I_6)
title("Emboss")

subplot(3, 4 ,7)
#I_1 = Image.open("nature1.jpg")
I_7 = I_1.filter(ImageFilter.FIND_EDGES)
plt.imshow(I_7)
title("Find Edges")

subplot(3, 4 ,8)
#I_1 = Image.open("nature1.jpg")
I_8 = I_1.filter(ImageFilter.SMOOTH)
plt.imshow(I_8)
title("Low Pass Filter(LPF1)")


subplot(3, 4 ,9)
#I_1 = Image.open("nature1.jpg")
I_9 = I_1.filter(ImageFilter.SMOOTH_MORE)
plt.imshow(I_9)
title("LPF2")


subplot(3, 4 ,10)
#I_1 = Image.open("nature1.jpg")
I_10 = I_1.filter(ImageFilter.SHARPEN)
plt.imshow(I_10)
title("Sharpen")

#Custom Kernels

size = (3, 3)
kernel1 = [1,1,1,0,0,0,-1,-1,-1]
ker1 = ImageFilter.Kernel(size,kernel1, scale=None, offset =0)
subplot(3,4,11)
I_11 = I_1.filter(ker1)
plt.imshow(I_11)
title("Custom Filter 1")


kernel2 = [1, 0, -1, 1, 0, -1, 0, 0, -1]
ker2 = ImageFilter.Kernel(size,kernel2,scale= None, offset =0)
subplot(3, 5, 12)
I_12 = I_1.filter(ker2)
plt.imshow(I_12)
title("Custom Filter 2")



```
